```
import os
os.chdir("/content/drive/My Drive/flipkart")

file = open("training.csv","r")
output = open("train_64.csv","w")
cnt = []
cnt.append(0)
cnt.append(0)
# cnt[0] and cnt[1] denote the x and y coordinates of the image respectively
# and h is the height of the image (difference in between the x coordinates, scaled by 480)
# and w is the width of the image (difference in y coordinates, scaled by 640)
fl = 0
for line in file:
  if fl==1:
    tmp = line.split(',')
    tmp[1] = int(int(tmp[1])*64/640)
    tmp[2] = int(int(tmp[2])*64/640)
    tmp[3] = int(int(tmp[3])*64/480)
    tmp[4] = int(int(tmp[4][:-1])*64/480)
    cnt[0] = int((tmp[3] + tmp[4])/2)
    cnt[1] = int((tmp[1] + tmp[2])/2)
    h = tmp[4] - tmp[3];
    w = tmp[2] - tmp[1];
#     print(cnt[0],cnt[1],h,w)
    prt = tmp[0] + "," + str(cnt[0]) + "," + str(cnt[1]) + "," + str(h) + "," + str(w) + "\n";
#     print(prt)
    output.write(prt)
  fl = 1
output.close()
```
